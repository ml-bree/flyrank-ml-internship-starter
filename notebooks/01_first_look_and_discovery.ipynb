## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
# ============================================================
# YOUR TURN: Discovery C - avg_position vs CTR
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt

# 1. Check correlation
correlation = df['avg_position'].corr(df['ctr'])
print(f"📊 Correlation between avg_position and CTR: {correlation:.3f}")

# 2. Create position tiers
df['position_tier'] = pd.cut(
    df['avg_position'],
    bins=[0, 3, 10, float('inf')],
    labels=['Top (1-3)', 'Middle (4-10)', 'Bottom (11+)']
)

# 3. Calculate average CTR by tier
tier_ctr = df.groupby('position_tier')['ctr'].mean().round(4)
print("\n📈 Average CTR by Position Tier:")
print(tier_ctr)

# 4. Visualize
tier_ctr.plot(kind='bar', color=['green', 'yellow', 'red'])
plt.title('CTR by Position Tier')
plt.ylabel('Average CTR')
plt.xlabel('Position Tier')
plt.xticks(rotation=0)
plt.show()

# 5. Optional: Check if there's a trend with content age
age_correlation = df['content_age_days'].corr(df['ctr'])
print(f"\n📅 Correlation between content_age_days and CTR: {age_correlation:.3f}")
```

## My Discovery: avg_position vs CTR

The correlation between avg_position and CTR was -0.073, which is weaker than expected.

However, when grouped into position tiers:
- Top positions (1-3): 2.71% average CTR
- Middle positions (4-10): 0.65% average CTR  
- Bottom positions (11+): 0.26% average CTR

Pages in the top 3 positions get ~4x more clicks than middle positions.

The correlation between content_age_days and CTR was 0.009, suggesting content age has minimal relationship with click-through rates.

**Insight:** The relationship isn't linear – the drop-off happens between tiers rather than smoothly. This is directional, not causal.
